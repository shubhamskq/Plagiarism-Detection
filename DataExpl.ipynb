```python
import numpy as np
import pandas as pd
import os
```

```python
myfile = 'data/Infofile.csv'
plagdf = pd.read_csv(myfile)
plagdf.head(10)
```

```python
print('Number of data files: ', plagdf.shape[0])
print('Number of unique tasks: ', (len(plagdf['Assctask'].unique())))
print('Unique plagscan categories: ', (plagdf['Cattype'].unique()))
```

```python
taskcount = plagdf.groupby(['Assctask']).size().reset_index(name="Counttask")
print("\nAssctask:")
display(taskcount)


categorycount = plagdf.groupby(['Cattype']).size().reset_index(name="Counttask")
print("\nPlagiarism Levels:")
display(categorycount)


task_plag_count = plagdf.groupby(
    ['Assctask', 'Cattype']).size().reset_index(name="Counttask")
print("\nAssctask & Plagiarism Levels :")
display(task_plag_count)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
group = ['Assctask', 'Cattype']
counts = plagdf.groupby(group).size().reset_index(name="Counttask")
plt.figure(figsize=(5, 3))
plt.bar(range(len(counts)), counts['Counttask'], color='red')
```
